```python
import matplotlib.pyplot as plt
import numpy as np
#create figure


X=np.arange(1/2,1,0.0001)
Y= 3*(1-X)/(2-X) #calculate the square of at each x
plt.plot(X,Y,label="Ingham")
X=np.arange(5/8,1,0.0001)
Y= (5*X-3)*(1-X)/(X*X+X-1)
plt.plot(X,Y,label="Huxley")
plt.xlabel('\\sigma')
plt.ylabel('Exponent')
plt.legend()
plt.title('Zero Density exponent for Ingham and Huxley Bounds')
#plt.savefig('coordinate_geometry/square_graph.png')
plt.savefig("inghamhuxley1.png",bbox_inches='tight')
```
