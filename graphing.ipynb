```python
import matplotlib.pyplot as plt
import numpy as np
#create figure


X=np.arange(1/2,1,0.0001)
Y= 3*(1-X)/(2-X) #calculate the square of at each x
plt.plot(X,Y,label="Ingham")
X=np.arange(5.3/8,1,0.0001)
Y= (3)*(1-X)/(3*X-1)
plt.plot(X,Y,label="Huxley")
plt.xlabel(fr'$\sigma$')
plt.ylabel('Exponent')
plt.legend()
plt.title('Zero Density exponent for Ingham and Huxley Bounds')
#plt.savefig('coordinate_geometry/square_graph.png')
plt.savefig("inghamhuxley1.png",bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import numpy as np
#create figure


X=np.arange(1/2,1,0.0001)
Y= 3*(1-X)/(2-X)/(1-X) #calculate the square of at each x
plt.plot(X,Y,label="Ingham")
X=np.arange(5.3/8,1,0.0001)
Y= (3)*(1-X)/(3*X-1)/(1-X)
plt.plot(X,Y,label="Huxley")
plt.xlabel(r'$\sigma$')
plt.ylabel(r'$\frac{Exponent}{(1-\sigma)}$')
plt.legend()
plt.title('Zero Density exponent for Ingham and Huxley Bounds')
#plt.savefig('coordinate_geometry/square_graph.png')
plt.savefig("inghamhuxley2.png",bbox_inches='tight')
```
